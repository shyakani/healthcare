```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/healthcare_dataset.csv')

df['Age'] = pd.to_numeric(df['Age'], errors='coerce')

emergency = df[(df['Gender'] == 'Female') & (df['Admission Type'] == 'Emergency')]['Age'].to_numpy()
routine = df[(df['Gender'] == 'Female') & (df['Admission Type'] == 'Routine')]['Age'].to_numpy()
urgent = df[(df['Gender'] == 'Female') & (df['Admission Type'] == 'Urgent')]['Age'].to_numpy()
```

```python
means = [
    df[(df['Gender'] == 'Female') & (df['Admission Type'] == 'Emergency')]['Age'].mean(),
    df[(df['Gender'] == 'Female') & (df['Admission Type'] == 'Routine')]['Age'].mean(),
    df[(df['Gender'] == 'Female') & (df['Admission Type'] == 'Urgent')]['Age'].mean()
]

print(means)
```

```python
s_t, sb, sd = [
    df[(df['Gender'] == 'Female') & (df['Admission Type'] == 'Emergency')]['Age'].var(ddof=1),
    df[(df['Gender'] == 'Female') & (df['Admission Type'] == 'Routine')]['Age'].var(ddof=1),
    df[(df['Gender'] == 'Female') & (df['Admission Type'] == 'Urgent')]['Age'].var(ddof=1)
]

print(s_t, sb, sd)
```

```python
# Define t, b, and d
t = df[(df['Gender'] == 'Female') & (df['Admission Type'] == 'Emergency')]['Age']
b = df[(df['Gender'] == 'Female') & (df['Admission Type'] == 'Routine')]['Age']
d = df[(df['Gender'] == 'Female') & (df['Admission Type'] == 'Urgent')]['Age']
nt, nb, nd = t.size, b.size, d.size
print(nt, nb, nd)
```

```python
import numpy as np

# Define the CIerr_calc function with a zero-check for sample size
def CIerr_calc(z, s_squared, n):
    if n == 0:
        return np.nan  # You can return None or 0 if preferred, np.nan indicates no valid CI
    return z * np.sqrt(s_squared / n)

# Z-score for 95% confidence interval
z = 1.96

# Calculate the CIerrs, ensuring sample size isn't zero
CIerrs = [
    CIerr_calc(z, s_t, nt),
    CIerr_calc(z, sb, nb),
    CIerr_calc(z, sd, nd)
]

# Print the CI errors (handling the np.nan result)
print(CIerrs)
```

```python
fig, ax = plt.subplots()
plt.title('Healthcare')
plt.ylabel('Weight (kg)')
plt.grid(axis='y')
_ = ax.errorbar(['Medical Condition', 'Blood Type', 'Medication'], means, CIerrs,
                fmt='o', color='green')
```
